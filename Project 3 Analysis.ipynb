```python
# import dependencies
import pandas as pd
```

```python
# read in csv file
file = "Resources/listings.csv"
df = pd.read_csv(file)
df.head()
```

```python
# display columns in df
df.columns
```

```python
# drop columns from df
newDF = df.drop(columns = ['scrape_id', 'last_scraped', 'source','description', 'neighborhood_overview', 'picture_url',\
                  'host_url', 'host_name', 'host_since', 'host_location', 'host_about','host_thumbnail_url', 'host_picture_url',\
                  'host_verifications','host_has_profile_pic', 'host_identity_verified','bathrooms','minimum_nights',\
                   'maximum_nights', 'minimum_minimum_nights','maximum_minimum_nights', 'minimum_maximum_nights','maximum_maximum_nights',\
                    'minimum_nights_avg_ntm','maximum_nights_avg_ntm', 'calendar_updated','neighbourhood_group_cleansed','calendar_last_scraped',\
                   'number_of_reviews_ltm', 'number_of_reviews_l30d', 'first_review','last_review','review_scores_accuracy',\
                   'license', 'instant_bookable','calculated_host_listings_count','calculated_host_listings_count_entire_homes',\
                   'calculated_host_listings_count_private_rooms','calculated_host_listings_count_shared_rooms', 'reviews_per_month','host_neighbourhood',\
                    'host_response_time', 'host_response_rate', 'host_acceptance_rate','neighbourhood','host_listings_count','beds'
                  ])
```

```python
# display df
newDF.head()
```

```python
# check data types
newDF.info()
```

```python
# drop null values
newDF = newDF.dropna()
```

```python
# change data type of super host from object to boolean
newDF['host_is_superhost'] = newDF['host_is_superhost'].map({'t': True, 'f': False}) 
```

```python
# import regex to remove letters from bathroom_text
import re
newDF['bathrooms_text'] = newDF['bathrooms_text'].str.replace(r'\D+','0', regex=True)
```

```python
# fill blank values
newDF.fillna(method='pad')
```

```python
# drop null values
newDF = newDF.dropna()
```

```python
# Convert data type to float
newDF['bathrooms_text'] = newDF['bathrooms_text'].astype(float)
newDF.rename(columns = {'bathrooms_text':'bathrooms'}, inplace = True)
```

```python
newDF.rename(columns = {'neighbourhood_cleansed':'county'},inplace=True)
```

```python
# convert price to float
newDF['price'] = newDF['price'].str.replace(',', '',regex=True).str.replace('$','',regex=True)
```

```python
newDF['price'] = newDF['price'].astype(float)
```

```python
# change data type of 'has_avaliability' from object to boolean
newDF['has_availability'] = newDF['has_availability'].map({'t': True, 'f': False}) 
```

```python
# Checking info type
newDF.info()
```
